```python
countries = f500["country"].unique()
```

```python
top_comp = {}

for c in countries:
    df = f500[f500["country"] == c]
    company = df.sort_values("employees", ascending = False).iloc[0]["company"]
    top_comp[c] = company
```

```python
print(top_comp )
```
